## 1. Konfigurasi

```python
# Auth token dari cookie X (DevTools > Application > Cookies > auth_token)
TWITTER_AUTH_TOKEN = "YOUR_AUTH_TOKEN_HERE"

# Kategori keyword: 'mbg' atau 'makan_bergizi_gratis'
KEYWORD_CATEGORY = "mbg"
SEARCH_QUERY     = "mbg since:2025-11-06 until:2025-11-09 lang:id"
FILENAME         = "6-8-nov-25.csv"
LIMIT            = 5000

assert KEYWORD_CATEGORY in {"mbg", "makan_bergizi_gratis"}
assert TWITTER_AUTH_TOKEN != "YOUR_AUTH_TOKEN_HERE"
```

## 2. Mount Drive

```python
import os
from google.colab import drive

drive.mount("/content/drive")

DRIVE_ROOT = "/content/drive/MyDrive/mbg_bert_cnn"
RAW_DIR    = os.path.join(DRIVE_ROOT, "dataset", "raw", KEYWORD_CATEGORY)
os.makedirs(RAW_DIR, exist_ok=True)

print(f"Output: {RAW_DIR}")
```

## 3. Install Node.js dan tweet-harvest

```python
!sudo apt-get update -qq
!sudo apt-get install -y ca-certificates curl gnupg -qq
!sudo mkdir -p /etc/apt/keyrings
!curl -fsSL https://deb.nodesource.com/gpgkey/nodesource-repo.gpg.key \
    | sudo gpg --dearmor -o /etc/apt/keyrings/nodesource.gpg
!echo "deb [signed-by=/etc/apt/keyrings/nodesource.gpg] https://deb.nodesource.com/node_20.x nodistro main" \
    | sudo tee /etc/apt/sources.list.d/nodesource.list
!sudo apt-get update -qq
!sudo apt-get install -y nodejs -qq
!npx playwright install-deps
!node -v
```

## 4. Crawling

```python
!npx -y tweet-harvest@2.7.1 \
    -o "{FILENAME}" \
    -s "{SEARCH_QUERY}" \
    --tab "LATEST" \
    -l {LIMIT} \
    --token {TWITTER_AUTH_TOKEN}
```

## 5. Verifikasi dan Simpan

```python
import shutil
import pandas as pd

LOCAL_PATH = os.path.join("tweets-data", FILENAME)
DRIVE_PATH = os.path.join(RAW_DIR, FILENAME)

assert os.path.exists(LOCAL_PATH), f"File tidak ditemukan: {LOCAL_PATH}"

df = pd.read_csv(LOCAL_PATH)
print(f"Jumlah tweet: {len(df):,}")

shutil.copy(LOCAL_PATH, DRIVE_PATH)
print(f"Tersimpan: {DRIVE_PATH}")
```
