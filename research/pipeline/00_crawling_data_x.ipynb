```python
# Crawl Data

filename = '8-nov-25.csv'
search_keyword = 'mbg since:2025-11-6 until:2025-11-09 lang:id'
limit = 5000

!npx -y tweet-harvest@2.7.1 -o "{filename}" -s "{search_keyword}" --tab "LATEST" -l {limit} --token {twitter_auth_token}
```
