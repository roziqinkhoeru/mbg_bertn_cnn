```python
# =========================
# 1. Mount Google Drive
# =========================
from google.colab import drive
drive.mount('/content/drive')

# =========================
# 2. Import Library
# =========================
import pandas as pd

# =========================
# 3. Path File
# =========================
file1 = "/content/drive/My Drive/skripsi/dataset/mbg/merged_data/merged_mbg.csv"
file2 = "/content/drive/My Drive/skripsi/dataset/mbg/merged_data/merged_makan_bergizi_gratis.csv"

output_path = "/content/drive/My Drive/skripsi/dataset/mbg/merged_data/merged_mbg_full.csv"

# =========================
# 4. Load Dataset
# =========================
df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)

# =========================
# 5. FIX KOLOM USERNAME
# =========================
username_cols = ['username', 'username\r', 'username;;;;;;;\r']

def merge_username(df):
    # Pastikan semua kolom tersedia
    for col in username_cols:
        if col not in df.columns:
            df[col] = None

    # Fungsi memilih username terbaik
    def pick_username(row):
        for col in ['username', 'username\r', 'username;;;;;;;\r']:
            val = row[col]
            if pd.notna(val) and str(val).strip() != "":
                return val
        return None

    # Apply ke dataframe
    df['username'] = df.apply(pick_username, axis=1)

    # Drop kolom rusak selain 'username'
    df = df.drop(columns=[col for col in username_cols if col != 'username'], errors='ignore')

    return df

df1 = merge_username(df1)
df2 = merge_username(df2)

# =========================
# 6. Tambahkan Source Keyword
# =========================
df1["source_keyword"] = "mbg"
df2["source_keyword"] = "makan_bergizi_gratis"

print("Dataset 1 shape:", df1.shape)
print("Dataset 2 shape:", df2.shape)

# =========================
# 7. Gabungkan Dataset
# =========================
df_full = pd.concat([df1, df2], axis=0, ignore_index=True)

# =========================
# 8. Simpan Hasil
# =========================
df_full.to_csv(output_path, index=False)

print(f"\nFile berhasil disimpan di: {output_path}")

# =========================
# 9. INFORMASI DATASET
# =========================
print("\n===== INFORMASI DATASET =====")

# Kolom
print("\nKolom dalam dataset:")
print(df_full.columns.tolist())

# Info umum
print("\nInfo dataset:")
df_full.info()

# Total data
print("\nTotal jumlah data:", len(df_full))

# Missing values
print("\nJumlah missing values per kolom:")
print(df_full.isnull().sum())

# Distribusi source
print("\nDistribusi source_keyword:")
print(df_full["source_keyword"].value_counts())

# Cek duplikat
print("\nJumlah data duplikat:", df_full.duplicated().sum())

# =========================
# 10. Preview Data
# =========================
print("\n===== 5 DATA AWAL =====")
display(df_full.head(5))

print("\n===== 5 DATA AKHIR =====")
display(df_full.tail(5))
```
